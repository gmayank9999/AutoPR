# 00 — Exploratory Data Analysis

Run this after `python src/preprocess.py` to explore the dataset and generate figures for the report.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

def load(name):
    return pd.DataFrame([
        json.loads(l) 
        for l in open(f"data/processed/{name}.jsonl", encoding="utf-8")
    ])

train = load("train")
val   = load("val")
test  = load("test")

print(f"Train: {len(train):,}  Val: {len(val):,}  Test: {len(test):,}")
print(f"Source mean/median: {train.src_len_words.mean():.1f} / {train.src_len_words.median():.1f}")
print(f"Target mean/median: {train.tgt_len_words.mean():.1f} / {train.tgt_len_words.median():.1f}")
print(f"Avg commits/PR: {train.num_commits.mean():.2f}")
```

```python
Path("report/figs").mkdir(parents=True, exist_ok=True)

# Source length histogram
fig, ax = plt.subplots(figsize=(8, 4))
train.src_len_words.clip(upper=400).hist(bins=50, ax=ax, color='steelblue', edgecolor='white')
ax.set_title("Train: source length distribution (words, clipped at 400)")
ax.set_xlabel("Words")
ax.set_ylabel("Count")
ax.axvline(train.src_len_words.mean(), color='red', linestyle='--', label=f'mean={train.src_len_words.mean():.0f}')
ax.legend()
plt.tight_layout()
plt.savefig("report/figs/source_length_hist.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved source_length_hist.png")
```

```python
# Target length histogram
fig, ax = plt.subplots(figsize=(8, 4))
train.tgt_len_words.clip(upper=150).hist(bins=50, ax=ax, color='darkorange', edgecolor='white')
ax.set_title("Train: target length distribution (words, clipped at 150)")
ax.set_xlabel("Words")
ax.set_ylabel("Count")
ax.axvline(train.tgt_len_words.mean(), color='red', linestyle='--', label=f'mean={train.tgt_len_words.mean():.0f}')
ax.legend()
plt.tight_layout()
plt.savefig("report/figs/target_length_hist.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved target_length_hist.png")
```

```python
# Commits per PR distribution
fig, ax = plt.subplots(figsize=(8, 4))
train.num_commits.clip(upper=20).value_counts().sort_index().plot(kind='bar', ax=ax, color='seagreen', edgecolor='white')
ax.set_title("Train: number of commits per PR (clipped at 20)")
ax.set_xlabel("Number of commits")
ax.set_ylabel("PRs")
plt.tight_layout()
plt.savefig("report/figs/num_commits_dist.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved num_commits_dist.png")
```
